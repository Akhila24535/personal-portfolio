```python
# TASK MANAGEMENT APPLICATION
# Run this entire code in a Google Colab cell

!pip -q install flask pyngrok

from flask import Flask, request, redirect, render_template_string
from pyngrok import ngrok
import threading

app = Flask(__name__)

tasks = []

HTML = """
<!DOCTYPE html>
<html>
<head>
    <title>Task Management Application</title>
    <style>
        body{
            font-family: Arial, sans-serif;
            background:#f4f4f4;
            padding:20px;
        }
        .container{
            width:60%;
            margin:auto;
            background:white;
            padding:20px;
            border-radius:10px;
            box-shadow:0px 0px 10px gray;
        }
        h1{
            text-align:center;
            color:#333;
        }
        input[type=text]{
            width:70%;
            padding:10px;
        }
        button{
            padding:10px;
            background:#007bff;
            color:white;
            border:none;
            border-radius:5px;
            cursor:pointer;
        }
        table{
            width:100%;
            margin-top:20px;
            border-collapse:collapse;
        }
        th,td{
            border:1px solid #ddd;
            padding:10px;
            text-align:center;
        }
        th{
            background:#007bff;
            color:white;
        }
        a{
            text-decoration:none;
            color:red;
        }
    </style>
</head>
<body>
<div class="container">
    <h1>Task Management Application</h1>

    <form action="/add" method="post">
        <input type="text" name="task" placeholder="Enter Task" required>
        <button type="submit">Add Task</button>
    </form>

    <table>
        <tr>
            <th>ID</th>
            <th>Task</th>
            <th>Action</th>
        </tr>

        {% for task in tasks %}
        <tr>
            <td>{{ loop.index }}</td>
            <td>{{ task }}</td>
            <td>
                <a href="/delete/{{ loop.index0 }}">Delete</a>
            </td>
        </tr>
        {% endfor %}
    </table>
</div>
</body>
</html>
"""

@app.route('/')
def home():
    return render_template_string(HTML, tasks=tasks)

@app.route('/add', methods=['POST'])
def add():
    task = request.form['task']
    tasks.append(task)
    return redirect('/')

@app.route('/delete/<int:index>')
def delete(index):
    if index < len(tasks):
        tasks.pop(index)
    return redirect('/')

def run():
    app.run(port=5000)

threading.Thread(target=run).start()

public_url = ngrok.connect(5000)
print("Open this URL in browser:")
print(public_url)
```
